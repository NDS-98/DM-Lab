```python
#Read the pima-indians-diabetes data file.
import numpy as np
import pandas as pd

from pandas import read_csv

dataset = read_csv('pima-indians-diabetes.data.csv',header=None)
dataset
```

```python
#Print the number of null values in the 7th column.
print(dataset.isnull().sum()[7])

dataset[7].isnull().sum()
```

```python
#Print the count of total number of null values in the dataset.
print(dataset.isnull().sum().sum())

#Replacing 0 by NaN and then counting total number of null values
dataset = dataset.replace(0,np.NaN);
print(dataset)

dataset.isnull().sum().sum()
```

```python
#Replace the missing values with the median of the respective columns.
dataset.fillna(dataset.mean(),inplace=True)

print(dataset.isnull().sum())

dataset
```

```python
#Print the count of total number of rows having missing values.
print(sum([True for idx,row in dataset.iterrows() if any (row.isnull())]))
```

```python
#Print the count of number of cells in the rows 120 to 198.
dataset[120:199].shape[0]*dataset[120:199].shape[1]
```

```python
#Create any dataframe and sample out 30% values from the dataframe.
df = pd.DataFrame({'Hello':[1,2,3,4], 'World':[5,6,7,8]}, index=['Any','Random','Value','Here'])
print(df)

df.sample(frac=0.3,replace=True,random_state=None)
```

```python
#Create a dataframe having 4 columns and 4 rows and having random values.Normalize it using min max normalisation.
df2 = pd.DataFrame(np.random.rand(4,4),columns=list('ABCD'),index=list('1234'))
print(df2)

normalized_df = (df2-df2.min())/(df2.max()-df2.min())
print(normalized_df)
```

```python
#Create a 2d numpy array of size (4,4)having 1 in first row ,2 in second row and so on.Find the correlation coefficient.
arr = np.random.rand(4,4)
arr[0]=1
arr[1]=2
arr[2]=3
arr[3]=4
print(arr)

np.corrcoef(arr,y=None,rowvar=True)
```

```python
#Create a numpy array having values 4,2,7,23,6.Print the interval corresponding to each value where first interval is 0 to 2
#,second interval is 2 to 4 and so on. 
arr2 = np.array([4,2,7,23,6])
print(arr2)
bins = np.array([i for i in range(0,25,2)])
print(bins)
inds = np.digitize(arr2,bins,right=False)
inds
```
