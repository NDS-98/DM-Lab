```python
from pandas import read_csv
import numpy as np
import pandas as pd
#READING THE DATA
dataset = read_csv('pima-indians-diabetes.data.csv', header=None)
print(dataset.describe())
```

```python
#Printing the first 20 rows of the dataset
print(dataset.head(20))
#Similarly tail is used to print the rows from the last.
```

```python
#Printing the number of zeroes in the respective columns
print((dataset[[1,2,3,4,5]] == 0).sum())
```

```python
#printing the number of null values in a dataset
print(dataset.isnull().sum())
```

```python
#Replacing the zeroes with Nan values in numpy
#Nan stands for not a number
dataset[[1,2,3,4,5]] = dataset[[1,2,3,4,5]].replace(0, np.NaN)
dataset
```

```python
#Gives the count of total number of rows having missing values 
print(sum([True for idx,row in dataset.iterrows() if any(row.isnull())]))

#Gives the count of total number of cells having missing values
sum(dataset.isnull().values.ravel())
```

```python
#printing the number of null values in a dataset
print(dataset.isnull().sum())

dataset.isnull().sum().sum()
```

```python
#removing missing values rows
dataset.dropna(inplace=True)

dataset
```

```python
#printing the dimensions of the dataset
print(dataset.shape)
```

```python
#replacing missing values rows
dataset.fillna(dataset.mean(), inplace=True)
print(dataset.isnull().sum())
print(dataset.head(20))

dataset
```

```python
#Sampling in pandas
df = pd.DataFrame({'num_legs': [2, 4, 8, 0],'num_wings': [2, 0, 0, 0],
                   'num_specimen_seen': [10, 2, 1, 8]},index=['falcon', 'dog', 'spider', 'fish'])
print(df)

print()
df['num_legs'].sample(n=3, random_state=None)#You can give it a specific state also by assigning numbers to it
```

```python
#sampling in pandas
df.sample(frac=0.5, replace=True, random_state=None)
```

```python
#Normalisation of columns of a dataframe
normalized_df=(df-df.mean())/df.std()
normalized_df
```

```python
#Min-Max normalisation
normalized_df=(df-df.min())/(df.max()-df.min())
normalized_df
```

```python
#digitisation using bin's index
x = np.array([0.2, 1.4, 3.0, 1.6])
print(x)
bins = np.array([0.0, 1.0, 2.5, 4.0, 10.0])
print(bins)
inds = np.digitize(x, bins,right=False)
inds
```
